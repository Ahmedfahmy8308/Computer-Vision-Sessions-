```python
import cv2
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
image = imread("photo/bug_noisy.jpg")
```

```python
def avg_blur(image):
    # The Kernel Side Length Must be Odd

    img1 = cv2.blur(image, (3, 3))
    img2 = cv2.blur(image, (5, 5))
    img3 = cv2.blur(image, (7, 7))
    img4 = cv2.blur(image, (13, 13))
    f , (img , im1,im2,im3,im4) = plt.subplots(1 , 5 ,figsize = (20 , 10))
    img.imshow(image)
    im1.imshow(img1)
    im2.imshow(img2)
    im3.imshow(img3)
    im4.imshow(img4)


```

```python
def gauss_blur(image):
    # By setting the third parameter to 0, we are instructing OpenCV
    # to automatically compute the weights based on our kernel size
    img1 = cv2.GaussianBlur(image, (3, 3),0)
    img2 = cv2.GaussianBlur(image, (5, 5),0)
    img3 = cv2.GaussianBlur(image, (7, 7),0)
    img4 = cv2.GaussianBlur(image, (13, 13),0)
    f , (img , im1,im2,im3,im4) = plt.subplots(1 , 5 ,figsize = (20 , 10))
    img.imshow(image)
    im1.imshow(img1)
    im2.imshow(img2)
    im3.imshow(img3)
    im4.imshow(img4)
```

```python
def median_blur(image):
    img1 = cv2.medianBlur(image, (3))
    img2 = cv2.medianBlur(image, (5))
    img3 = cv2.medianBlur(image, (7))
    img4 = cv2.medianBlur(image, (13))
    f , (img , im1,im2,im3,im4) = plt.subplots(1 , 5 ,figsize = (20 , 10))
    img.imshow(image)
    im1.imshow(img1)
    im2.imshow(img2)
    im3.imshow(img3)
    im4.imshow(img4)


```

```python
def bilateral(image):
    img1 = cv2.bilateralFilter(image, 3 , 21 ,21)
    img2 = cv2.bilateralFilter(image, 5,31,31)
    img3 = cv2.bilateralFilter(image, 7,41,41)
    img4 = cv2.bilateralFilter(image, 12,51,51)
    f , (img , im1,im2,im3,im4) = plt.subplots(1 , 5 ,figsize = (20 , 10))
    img.imshow(image)
    im1.imshow(img1)
    im2.imshow(img2)
    im3.imshow(img3)
    im4.imshow(img4)


```

```python
avg_blur(image)
gauss_blur(image)
median_blur(image)
bilateral(image)
```
